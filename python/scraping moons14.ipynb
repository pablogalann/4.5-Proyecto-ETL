```python
def limpiar_fechas(x):
    
    if 'de' in x:
        x = x.split(' de ')
        if len(str(x[0]))==1:
            x = x[1]+'0'+x[0]
        else:
            x = x[1]+x[0]

    if 'ene' in x:
        x ='2014/'+ x.replace('ene','01/')
    if 'feb' in x:
        x ='2014/'+ x.replace('feb','02/')
    if 'mar' in x:
        x ='2014/'+ x.replace('mar','03/')
    if 'abr' in x:
        x ='2014/'+ x.replace('abr','04/')
    if 'may' in x:
        x ='2014/'+ x.replace('may','05/')
    if 'jun' in x:
        x ='2014/'+ x.replace('jun','06/')
    if 'jul' in x:
        x ='2014/'+ x.replace('jul','07/')
    if 'ago' in x:
        x ='2014/'+ x.replace('ago','08/')
    if 'sep' in x:
        x ='2014/'+ x.replace('sep','09/')
    if 'oct' in x:
        x ='2014/'+ x.replace('oct','10/')
    if 'nov' in x:
        x ='2014/'+ x.replace('nov','11/')
    if 'dic' in x:
        x ='2014/'+ x.replace('dic','12/')
        
    return(x)
    
```

```python
df14['new_moon']=df14['new_moon'].apply(limpiar_fechas)
df14['first_quarter']=df14['first_quarter'].apply(limpiar_fechas)
df14['full_Moon']=df14['full_Moon'].apply(limpiar_fechas)
df14['third_quarter']=df14['third_quarter'].apply(limpiar_fechas)
```

###  Conversion hora 

```python
def limpiar_horas(x):

    if ':' in x:
        x = x.split(':')
        if len(str(x[0]))==1:
            x =' '+'0'+x[0]+':'+x[1]+':'+'00'
                
        else:
            x =' '+x[0]+':'+x[1]+':'+'00'
                
            
    return(x)  
```

```python
df14['time_new_moon']=df14['time_new_moon'].apply(limpiar_horas)
df14['time_first_quarter']=df14['time_first_quarter'].apply(limpiar_horas)
df14['time_full_moon']=df14['time_full_moon'].apply(limpiar_horas)
df14['time_third_quarter']=df14['time_third_quarter'].apply(limpiar_horas)
```
