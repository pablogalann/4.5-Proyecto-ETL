```python
def limpiar_hora(x):
    
    if ' ' in x:
        x=x.split(' ')
        
    
    return x[1]
```

```python
def limpiar_fecha(x):
    
    if ' ' in x:
        x=x.split(' ')
        
    
    return x[0]
```

```python
dlu['time_new_moon'] = dlu.loc[:, 'new_moon']
dlu['time_first_quarter'] = dlu.loc[:, 'first_quarter']
dlu['time_full_Moon'] = dlu.loc[:, 'full_Moon']
dlu['time_third_quarter'] = dlu.loc[:, 'third_quarter']
```

```python
dlu.columns
```

```python
new_colu=['id_ciclo_lunar', 'new_moon', 'time_new_moon','first_quarter', 'time_first_quarter', 'full_Moon','time_full_Moon',
       'third_quarter','time_third_quarter']
        
```

```python
dlu.head(5)
```

```python
dlu['new_moon']=dlu['new_moon'].apply(limpiar_fecha)
dlu['first_quarter']=dlu['first_quarter'].apply(limpiar_fecha)
dlu['full_Moon']=dlu['full_Moon'].apply(limpiar_fecha)
dlu['third_quarter']=dlu['third_quarter'].apply(limpiar_fecha)
```

```python
dlu['time_new_moon']=dlu['time_new_moon'].apply(limpiar_hora)
dlu['time_first_quarter']=dlu['time_first_quarter'].apply(limpiar_hora)
dlu['time_full_Moon']=dlu['time_full_Moon'].apply(limpiar_hora)
dlu['time_third_quarter']=dlu['time_third_quarter'].apply(limpiar_hora)
```

```python
dlu['new_moon'] = pd.to_datetime(dlu['new_moon'], errors='coerce')
dlu['first_quarter'] = pd.to_datetime(dlu['first_quarter'], errors='coerce')
dlu['full_Moon'] = pd.to_datetime(dlu['full_Moon'], errors='coerce')
dlu['third_quarter'] = pd.to_datetime(dlu['third_quarter'], errors='coerce')
```
