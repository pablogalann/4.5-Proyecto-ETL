```python
def limpiar_fechas(x):

    if '/' in x:
        x = x.split('/')
        for i in x:
            if int(x[1])<= 12:
                i = x[2]+'/'+x[1]+'/'+x[0]
            else:
                i = x[2]+'/'+x[0]+'/'+x[1]
            
    return(i)   
    
    
        
```

```python
dc['CrimeDate']=dc['CrimeDate'].apply(limpiar_fechas)
```
